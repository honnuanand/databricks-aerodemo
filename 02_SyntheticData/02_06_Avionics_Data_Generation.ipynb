```python
import os
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
import random

def generate_avionics_data(num_records_per_aircraft=100):
    aircraft_ids = ["A320_101", "A330_201", "B737_301"]
    data = {
        'avionics_id': [],
        'aircraft_id': [],
        'event_timestamp': [],
        'navigation_status': [],
        'autopilot_engaged': [],
        'sensor_health': [],
        'signal_strength': [],
        'software_version_drift': []
    }

    for aircraft_id in aircraft_ids:
        for i in range(num_records_per_aircraft):
            random_days_ago = random.randint(0, 6)
            random_time = datetime.now() - timedelta(
                days=random_days_ago,
                hours=random.randint(0, 23),
                minutes=random.randint(0, 59)
            )
            data['avionics_id'].append(f'AV_{aircraft_id}_{i:03d}')
            data['aircraft_id'].append(aircraft_id)
            data['event_timestamp'].append(random_time.strftime("%Y-%m-%d %H:%M:%S"))
            data['navigation_status'].append(random.choice(['OK', 'WARN', 'FAIL']))
            data['autopilot_engaged'].append(random.choice([True, False]))
            data['sensor_health'].append(round(np.random.uniform(0.8, 1.0), 2))  # normalized score
            data['signal_strength'].append(round(np.random.uniform(70, 100), 2))  # percentage
            data['software_version_drift'].append(round(np.random.uniform(0, 5), 2))  # versions behind

    df = pd.DataFrame(data)
    return df

# Generate DataFrame
df = generate_avionics_data()

# Save to Auto Loader-compatible path
output_path = "/Volumes/arao/aerodemo/tmp/avionics"
os.makedirs(output_path, exist_ok=True)
output_file = f"{output_path}/avionics_sample.csv"
if os.path.exists(output_file):
    os.remove(output_file)  # safely remove old file

df.to_csv(output_file, index=False)

print(f"âœ… Avionics data generated: {len(df)} rows saved to {output_file}")
```
